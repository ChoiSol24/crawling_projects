해당 프로그램은 chromedriver.exe 파일이 있어야 실행이 가능.
사용자의 크롬 버전에 맞는 chromedriver.exe 다운 후 실행 요망.

```python
import sys

from selenium import webdriver
import time
import random as r


from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import urllib.parse
from urllib.request import Request, urlopen

# 웹드라이버 켜기
driver = webdriver.Chrome(executable_path = './chromedriver.exe')

target = 'https://www.instagram.com/accounts/login/?source=auth_switcher'
driver.get(target)

#자동 로그인
time.sleep(2)
driver.find_elements_by_name("username")[0].send_keys('인스타 아이디 입력')
time.sleep(3)
p = driver.find_elements_by_name("password")[0]
p.send_keys("인스타 비밀번호 입력")
time.sleep(3)
p.submit()

#dm 진입 : 가끔 진입안되는 오류가 생길 수 있어 같은 주소 입력 한번 더 함
time.sleep(3)
driver.get('https://www.instagram.com/direct/inbox/')
time.sleep(3)
driver.get('https://www.instagram.com/direct/inbox/')
time.sleep(3)
WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.XPATH, '//button[text()="나중에 하기"]/..'))).click()

#dm 리스트 받아오기
El = driver.find_element_by_css_selector('div.N9abW')

#dm 리스트에서 id만 모은 리스트 만들기 및 출력
name = []
prev = ""
check = 0
while True:
    try:
        print(check)
        check+=1
        ans = driver.find_elements_by_css_selector('div._7UhW9.xLCgt.MMzan.KV-D4.fDxYl')
        time.sleep(2)
        print(ans[5].text)
        for k in ans :
            if k.text not in name :
                name.append(k.text)
        if prev == name[-1]:
            break
        prev = name[-1]
        print(name)
        time.sleep(2)
        #자동 스크롤
        driver.execute_script("arguments[0].scrollBy(0,800)", El)
        time.sleep(2)
        
    except:
        break

#공란 데이터 입력되는 것 빼기
if '' in name :
    name.remove('')

#name : dm 주요 메시지함의 아이디 목록 출력
print(name)
print('\n\n')

# 공통 - 파이썬 프로그램을 이용하여 랜덤으로 뽑을 시 사용
#winner : 당첨자 리스트
#r.sample(name, 뽑을 인원 수)
winner = r.sample(name,3)

#print(winner) : 등수 없이 뽑을 때

#winner의 순서대로 등수 결정
for i in range(3):
    print(winner[i],str(i+1)+"등입니다 축하드려요!")


#아이디 모은 것 엑셀 파일로 저장
import pandas as pd
df = pd.DataFrame(name)

df.to_excel('id_list.xlsx')

print('엑셀 파일 생성 완료')

```
